```python
# 7) Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by
 #subtracting only negative transaction_amounts for Relationship_type ="P" ( P - means  Primary , S - means Secondary ) .(27 Rows)

query="""
SELECT
    c.customer_id,
    a.Account_Number,
    a.Account_type,
    (a.Balance_amount + COALESCE(SUM(CASE WHEN t.Transaction_amount < 0 THEN t.Transaction_amount ELSE 0 END), 0)) AS deducted_balance_amount
FROM
    CUSTOMER1 c
JOIN
    BANK_ACCOUNT1 a ON c.customer_id = a.Customer_id
LEFT JOIN
    BANK_ACCOUNT_TRANSACTION1 t ON a.Account_Number = t.Account_Number
WHERE
    a.Relation_ship = 'P'
GROUP BY
    c.customer_id, a.Account_Number, a.Account_type, a.Balance_amount
ORDER BY
    c.customer_id, a.Account_Number;
"""
result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='psql'))
```

```python
# 10) Display all saving account holders have “Add-on Credit Cards" and “Credit cards" (3 Rows)
query="""
WITH AccountOwner AS (
    -- Level 0: Direct accounts from BANK_ACCOUNT1
    SELECT Customer_id, Account_Number, Account_type FROM BANK_ACCOUNT1
    UNION ALL
    -- Level 1: Accounts linked directly to BANK_ACCOUNT1 accounts
    SELECT
        ba.Customer_id,
        bard_l1.Account_Number,
        bard_l1.Account_type
    FROM
        Bank_Account_Relationship_Details1 bard_l1
    JOIN
        BANK_ACCOUNT1 ba ON bard_l1.Linking_Account_Number = ba.Account_Number
    WHERE
        bard_l1.Customer_id IS NULL
    UNION ALL
    -- Level 2: Accounts linked to Level 1 accounts (which are linked to BANK_ACCOUNT1 accounts)
    SELECT
        ba.Customer_id,
        bard_l2.Account_Number,
        bard_l2.Account_type
    FROM
        Bank_Account_Relationship_Details1 bard_l2
    JOIN
        Bank_Account_Relationship_Details1 bard_l1 ON bard_l2.Linking_Account_Number = bard_l1.Account_Number
    JOIN
        BANK_ACCOUNT1 ba ON bard_l1.Linking_Account_Number = ba.Account_Number
    WHERE
        bard_l2.Customer_id IS NULL AND bard_l1.Customer_id IS NULL
)
SELECT
    c.customer_id,
    c.customer_name,
    c.Address,
    c.state_code,
    c.Telephone
FROM
    CUSTOMER1 c
WHERE
    EXISTS (SELECT 1 FROM AccountOwner ao WHERE ao.Customer_id = c.customer_id AND ao.Account_type = 'SAVINGS')
AND
    EXISTS (SELECT 1 FROM AccountOwner ao WHERE ao.Customer_id = c.customer_id AND ao.Account_type IN ('CREDITCARD', 'Credit Card'))
AND
    EXISTS (SELECT 1 FROM AccountOwner ao WHERE ao.Customer_id = c.customer_id AND ao.Account_type = 'Add-on Credit Card');
"""
result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='psql'))
```

```python
# 11)  Display  records of “SAVINGS” account linked with “Credit cards" account_type and its credit
# aggregate sum of transaction amount. (1 Row)
# Ref: Check linking relationship in bank_transaction_relationship_details.
        # Check transaction_amount in bank_account_transaction.
query ="""
SELECT
    ba.Account_Number AS Savings_Account_Number,
    ba.Account_type AS Savings_Account_Type,
    bard.Account_Number AS Linked_Credit_Card_Account_Number,
    SUM(bat.Transaction_amount) AS Aggregate_Credit_Card_Transaction_Amount
FROM
    BANK_ACCOUNT1 ba
JOIN
    Bank_Account_Relationship_Details1 bard ON ba.Account_Number = bard.Linking_Account_Number
JOIN
    BANK_ACCOUNT_TRANSACTION1 bat ON bard.Account_Number = bat.Account_Number
WHERE
    ba.Account_type = 'SAVINGS'
    AND bard.Account_type IN ('CREDITCARD', 'Credit Card')
GROUP BY
    ba.Account_Number, ba.Account_type, bard.Account_Number;
"""
result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='psql'))
```
